# product management project

```python
from datetime import date

def manage_products():
    products = {}  # Dictionary to store product information

    while True:
        print("\n==== Product Management Menu ====")
        print("1. Add Product")
        print("2. Check Warranty")
        print("3. Get Location")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == "1":
            add_product(products)
        elif choice == "2":
            check_warranty(products)
        elif choice == "3":
            get_location(products)
        elif choice == "4":
            print("Exiting Product Management")
            break
        else:
            print("Invalid choice. Please try again.")

def add_product(products):
    name = input("Enter the product name: ")
    purchase_year = int(input("Enter the year of purchase: "))
    warranty_duration = int(input("Enter the warranty duration in years: "))
    location = input("Enter the product location: ")

    products[name] = {"purchase_year": purchase_year, "warranty_duration": warranty_duration, "location": location}
    print("Product added successfully.")

def check_warranty(products):
    name = input("Enter the product name: ")

    product = products.get(name)
    if product is not None:
        purchase_year = product["purchase_year"]
        warranty_duration = product["warranty_duration"]
        warranty_expiry_year = purchase_year + warranty_duration
        current_year = date.today().year

        warranty_status = "under warranty" if current_year <= warranty_expiry_year else "expired"
        print(f"The {name} product is {warranty_status}.")
    else:
        print("Product not found.")

def get_location(products):
    name = input("Enter the product name: ")

    if name in products:
        location = products[name]["location"]
        print(f"The {name} is located at: {location}")
    else:
        print("Product not found.")

manage_products()
```
